```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('hackathon_loan_data.csv')
df.columns
```

```python
filtered_columns = ['zip_code','Total_All_loans','White_All_loans','Black_All_loans','White_Unanswered_loans','Black_Unanswered_loans','White_All_loan_amnt','Black_All_loan_amnt']
df1 = df[filtered_columns]
```

```python
df1
```

```python
df1['loan_amount_per_person_white'] = df1.apply(lambda x: x['White_All_loan_amnt'] / x['White_All_loans'], axis=1)
df1['loan_amount_per_person_black'] = df1.apply(lambda x: x['Black_All_loan_amnt'] / x['Black_All_loans'], axis=1)
```

```python
df1
```

```python
filtered_columns = ['zip_code', 'loan_amount_per_person_white','loan_amount_per_person_black']
df2 = df1[filtered_columns]
```

```python
df2
```

```python
df2m = df2.melt(id_vars='zip_code')
sns.barplot(data=df2m, x='variable', y='value', hue='zip_code')
```
